```python
from selenium import webdriver
import pandas as pd

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")

driver = webdriver.Chrome(options=chrome_options)
```

```python
url = 'https://cryptoslate.com/cryptos/layer-1/'
driver.get(url)
```

```python
rank = driver.find_elements(by='xpath', value='//td[contains(@class, "rank")]')
name = driver.find_elements(by='xpath', value='//td[contains(@class, "name")]')
price = driver.find_elements(by='xpath', value='//td[contains(@class, "price")]')
chg_24h = driver.find_elements(by='xpath', value='//td[contains(@class, "change-24h")]')
chg_7d = driver.find_elements(by='xpath', value='//td[contains(@class, "change-7d")]')
chg_30d = driver.find_elements(by='xpath', value='//td[contains(@class, "change-30d")]')
market_cap = driver.find_elements(by='xpath', value='//td[contains(@class, "market-cap")]')
volume = driver.find_elements(by='xpath', value='//td[contains(@class, "volume")]')
```

```python
name_list = []
for x in range(len(name)):
  data = dict()

  data['rank'] = rank[x].text
  data['name'] = name[x].text
  data['price $'] = price[x].text
  data['chg_24h'] = chg_24h[x].text
  data['chg_7d'] = chg_7d[x].text
  data['chg_30d'] = chg_30d[x].text
  data['market_cap $'] = market_cap[x].text
  data['volume_24h $'] = volume[x].text

  name_list.append(data)
```

```python
df = pd.DataFrame(name_list)
df
```

```python
df['name'] = df['name'].str.split('\n').str[0]
df['price $'] = df['price $'].str.split('$')[0][1]
df['market_cap $'] = df['market_cap $'].str.split('$')[0][1]
df['volume_24h $'] = df['volume_24h $'].str.split('$')[0][1]
```

#Pagination Scrap

```python
for i in range(0,2):
  # Mencari element tag a dari ul forceShow
  consensus_ul = driver.find_element(by='xpath', value="//ul[contains(@class, 'forceShow')]")
  consensus_li = consensus_ul.find_elements(by='xpath', value=".//a")

  # Membuat url baru
  new_url = consensus_li[i].get_attribute('href')
  driver.get(new_url)

  # Value yang akan diambil
  rank = driver.find_elements(by='xpath', value='//td[contains(@class, "rank")]')
  name = driver.find_elements(by='xpath', value='//td[contains(@class, "name")]')
  price = driver.find_elements(by='xpath', value='//td[contains(@class, "price")]')
  chg_24h = driver.find_elements(by='xpath', value='//td[contains(@class, "change-24h")]')
  chg_7d = driver.find_elements(by='xpath', value='//td[contains(@class, "change-7d")]')
  chg_30d = driver.find_elements(by='xpath', value='//td[contains(@class, "change-30d")]')
  market_cap = driver.find_elements(by='xpath', value='//td[contains(@class, "market-cap")]')
  volume = driver.find_elements(by='xpath', value='//td[contains(@class, "volume")]')

  # Scraping data
  name_list = []
  for x in range(len(name)):
    data = dict()

    data['rank'] = rank[x].text
    data['name'] = name[x].text
    data['price $'] = price[x].text
    data['chg_24h'] = chg_24h[x].text
    data['chg_7d'] = chg_7d[x].text
    data['chg_30d'] = chg_30d[x].text
    data['market_cap $'] = market_cap[x].text
    data['volume_24h $'] = volume[x].text

    name_list.append(data)

  # Cleaning
  df = pd.DataFrame(name_list)
  df['name'] = df['name'].str.split('\n').str[0]
  df['price $'] = df['price $'].str.split('$')[0][1]
  df['market_cap $'] = df['market_cap $'].str.split('$')[0][1]
  df['volume_24h $'] = df['volume_24h $'].str.split('$')[0][1]

  # Save to csv
  table_name = new_url.split("/")[4]
  df.to_csv(f'{table_name}.csv')
```
